```python
import seaborn as sns
```

```python
# Lista de comandos possíveis 
sns.get_dataset_names()
```

```python
# Exemplos
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips
```

```python
# hue => é a matriz! Nesse caso deu coloração 
# ao gráfico para distiguir os dias
# size deu tamanhos diferentes
# YlGnBu - paleta yellow, green e blue
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette="YlGnBu")
```

```python
sns.histplot(tips.tip, kde=True, bins=15)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='YlGnBu', showfliers=False)
```

```python
# dodge = desviar. Dá um afastamento entre as informações.
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge=True)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='YlGnBu')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex')
```

```python
# Se o pré-processamento de dados tem que ser feito em Python, 
# então este comando pode salvar algum tempo. 
# Depois de ler a tabela, o tipo padrão de dados para cada coluna pode ser
# bool, int64, float64, object, category, timedelta64 ou datetime64. 
# Você pode checar antes, usando df.dtypes.value_counts() para saber
# os possíveis tipos de dados do seu dataframe, e então usar df.select_dtypes(include=['float64', 'int64']) 
# para selecionar um sub-dataframe apenas com features numéricas.
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```
