```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# norm.rvs - cria dados dados em uma distribuição normal
dados = norm.rvs(size = 1000)
dados
```

```python
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Verificando se a distribuição é normal 
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
# Teste Shapiro
# Maior que 0.05 não dá para rejeitar a hipótese 
stats.shapiro(dados)
```

```python
# Dados não normais 
# skewnorm - assimetria
dados2 = skewnorm.rvs(4, size=1000)
```

```python
plt.hist(dados2, bins=20)
plt.title('Dados')
```

```python
# Verificando se os dados são simétricos
fig,ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
# Verificando se a distribuição é normal pelo método Shapiro
# Acima de 0.05 não dá para descartar como nulo 
stats.shapiro(dados2)
```
