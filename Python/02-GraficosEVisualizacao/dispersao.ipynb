```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.tail()
```

```python
# facecolors = cor de fundo
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors='#8c564b', marker = '+')
plt.title('Árvores')
plt.xlabel('Circuferência')
plt.ylabel('Volume')
```

```python
plt.plot(base.Girth, base.Volume)
plt.title('Árvore')
plt.xlabel('Circurferência')
plt.ylabel('Volume')
```

```python
# fit_reg = linha de tendência 
# x_jitter afasta um pouco os dados iguais para serem melhor visualizados no gráfico.   
sns.regplot(x=base.Girth,y=base.Volume, data = base , x_jitter = 0.3, fit_reg=True, marker='*')
```
